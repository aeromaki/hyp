```python
import torch
from torch import nn, optim
from structures.functions import hyperboloid, dist_h
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
def dist(
    a: torch.Tensor,
    b: torch.Tensor
) -> torch.Tensor:
    ah = hyperboloid(a)
    bh = hyperboloid(b)
    return dist_h(ah, bh)

def criterion(
    embed: torch.Tensor,
    mask_pos: torch.Tensor,
    mask_neg: torch.Tensor
) -> torch.Tensor:
    d = dist(embed, embed)
    pos = d[mask_pos == 1]
    neg = d[mask_neg == 0]
    return pos.mean() + pos.var() - neg.mean() + neg.var()
```

```python
graph = graph_s
graph = torch.tensor(graph)
(n_label := graph.shape[0])
```

```python
d_model = 16
lr = 0.1 #e-03
n_iter = 30000
n_batch = 10000

model = nn.Parameter(torch.rand(n_label, d_model))

mask_pos = graph
mask_neg = mask_pos | torch.eye(n_label).type(torch.int)

optimizer = optim.Adam([model], lr=lr)
```

```python
s = 0
for i in tqdm(range(n_iter)):
    optimizer.zero_grad()
    embed = model
    loss = criterion(embed, mask_pos, mask_neg)
    loss.backward()
    optimizer.step()
    s += loss.item()
    if (i+1) % n_batch == 0:
        print(s / n_batch)
        s = 0
```
